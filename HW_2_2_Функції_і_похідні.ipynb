### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
```

```python
x = sp.symbols('x')
```

```python
f_score = 10*x - 5
derivative_f_score = sp.diff(f_score, x)
```

```python
xs = np.linspace(0, 10, 10)
f = 10*xs - 5
df = 10*xs**0

plt.figure(figsize=(10,10))
plt.plot(xs, f, label='f(x) = 10x - 2', color='blue')
plt.plot(xs, df, label=f"f'(x) = {derivative_f_score}", color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Зміна кількості годин вивчення лінійно впливає на оцінку, тобто із кожною наступною затраченою годиною оцінка буде збільшуваись на ту ж кількість балів, що і при попередньо затраченій годині.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)
```

```python
xs = np.linspace(0, 10, 100)
f_sin = np.sin(xs)
derivative_f_sin = np.cos(xs)

plt.figure(figsize=(10,3))
plt.plot(xs, f_sin, label='f(x) = sin(x)', color='blue')
plt.plot(xs, derivative_f_sin, label='f(x) = cos(x)', color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)
```

```python
xs = np.linspace(0.001, 0.03, 1000)
f_ln = np.log(xs)
derivative_f_ln = 1 / xs

plt.figure(figsize=(10,3))
plt.plot(xs, f_ln, label='f(x) = ln(x)', color='blue')
plt.plot(xs, derivative_f_ln, label='1/x', color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f_spend = x**2 - 5*x + 1
derivative_f_spend = sp.diff(f_spend, x)
```

```python
xs = np.linspace(0, 10, 1000)
f_spend = xs**2 - 5*xs + 1
derivative_f_spend = 2*xs - 5

plt.figure(figsize=(10,5))
plt.plot(xs, f_spend, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(xs, derivative_f_spend, label='2x - 5', color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
